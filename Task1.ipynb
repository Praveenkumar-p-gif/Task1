```python
import pandas as pd
```

```python
df = pd.read_csv('/content/netflix_show.csv')
df.head(10)
```

```python
#Identify and handle missing values using .isnull() in Python or filters in Excel.
df.isnull().sum()
```

```python
##Remove duplicate rows using .drop_duplicates() or Excel’s “Remove Duplicates”.
##Standardize the column names by removing spaces and converting to lowercase.

df = df.drop_duplicates()
print(df)
```

```python
##Standardize text values like gender, country names, etc.
if 'country' in df.columns:
    df['country'] = df['country'].str.strip().str.title()
    print(df['country'])
```

```python
##Convert date formats to a consistent type (dd-mm-yyyy)
if 'date_added' in df.columns:
    df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce').dt.strftime('%d-%m-%Y')
    print(df['date_added'])
```

```python
##Rename column headers (lowercase, replace spaces with underscores)
df.columns = df.columns.str.lower().str.strip().str.replace(' ', '_')
print(df.columns)
```

```python
##Check and fix data types
if 'release_year' in df.columns:
    df['release_year'] = pd.to_numeric(df['release_year'], errors='coerce').astype('Int64')
print(df['release_year'])
```
